```python
#er_pr#lets create the new variables
def combine_er_pr(er, pr):
    if er == 'Negative' and pr == 'Negative':
        return 'Double Negative'
    elif er == 'Positive' and pr == 'Negative':
        return 'ER+/PR-'
    elif er == 'Negative' and pr == 'Positive':
        return 'ER-/PR+'
    elif er == 'Positive' and pr == 'Positive':
        return 'Double Positive'
    else:
        return None
brc['er_pr'] = brc.apply(lambda row: combine_er_pr(row['er'], row['pr']), axis=1)
brc.head()
print(brc.dtypes)
#er_pr 
erprcounts=brc['er_pr'].value_counts()

fig=plt.figure()
plt.bar(erprcounts.index,erprcounts.values)
plt.title('Hormone Receptor Positivity')
plt.xlabel('Category')
plt.ylabel('count')
plt.show()
```

### Association tests:

```python
#now take a look at some association tests
from scipy.stats import chi2_contingency
# Create crosstab
ct1 = pd.crosstab(brc['Status'], brc['MaritalStatus'])

# Run chi-square test
chi2_1, p_1, dof_1, expected_1 = chi2_contingency(ct1)

print("MaritalStatus association with Status")
print("Chi2:", chi2_1, "| p-value:", p_1)
print(ct1)

# Add totals (margins)
ct1_with_totals = pd.crosstab(brc['Status'], brc['MaritalStatus'], margins=True)
print("\nCounts with totals:")
print(ct1_with_totals)



```

```python
#Now lets run it again
# Create crosstab
ct2 = pd.crosstab(brc['Status'], brc['married_status'])

# Run chi-square test
chi2_2, p_2, dof_2, expected_2 = chi2_contingency(ct2)

print("MaritalStatus association with Status")
print("Chi2:", chi2_2, "| p-value:", p_2)
print(ct2)

# Add totals (margins)
ct2_with_totals = pd.crosstab(brc['Status'], brc['married_status'], margins=True)
print("\nCounts with totals:")
print(ct2_with_totals)
```

#### There is no association with Marital status before combining the categories, now there is after combining.

```python
#now lets do ER/PR 

#lets create the new variables
def erprstatus(er,pr,):
    if er == 'Positive':
        return 1
    elif pr == 'Positive':
        return 1
    elif er == 'Negative' and pr == 'Negative':
        return 0
    else:
        return None  # handles missing or unexpected values
brc['hormone_therapy'] = brc.apply(lambda row: erprstatus(row['er'], row['pr']), axis=1)
brc.head()
```

```python
#Now lets run the association test
# Create crosstab
ct3 = pd.crosstab(brc['Status'], brc['hormone_therapy'])

# Run chi-square test
chi2_3, p_3, dof_3, expected_3 = chi2_contingency(ct3)

print("HOrmone therapy association with Status")
print("Chi2:", chi2_3, "| p-value:", p_3)
print(ct3)

# Add totals (margins)
ct3_with_totals = pd.crosstab(brc['Status'], brc['hormone_therapy'], margins=True)
print("\nCounts with totals:")
print(ct3_with_totals)
```

#### Association between hormone therapy with live status
